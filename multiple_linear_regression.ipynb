```python
#a multiple linear regression attempts to model relatioship between two or more explanatory variable

#The equation can be moulded as y = b0 + b1X1 + b2X2 + .... + bnXn
```

```python
#data preprocessing part

#importing libraries
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd

#import dataset
dataset = pd.read_csv("C:/Users/Nill/Desktop/Machine Learning A-Z/Part 2 - Regression/Section 5 - Multiple Linear Regression/50_Startups.csv")
dataset.head()
```

```python
#we create independent and dependent variable vector or feature matrix

#independet variable (number of years) or matrix of features (-1 removes last column of the dataset)
X = dataset.iloc[:, :-1].values

#get the dependent variable (salary) that has an index of 1
Y = dataset.iloc[:, 4].values
```

```python
print(X)
```

```python
#Encoding categorical Data

#the dataset, country contains catgeories (New York, floride, california)
#we encode text in above categories as numbers

#library for encoding
from sklearn.preprocessing import LabelEncoder
labelencoder_X = LabelEncoder()

#apply labeEncoder object on our column (3 is the index of categorical column)
X[:,3] = labelencoder_X.fit_transform(X[:,3])

#output contains country as encoded values
print(X)
```

```python
#categorical encoding could lead us to wrong results as it can imply that one value could be greater than the other(not in this case)

#dummy encoding with oneHotEncoder
from sklearn.preprocessing import OneHotEncoder

#specify which column should be onehotencode category
onehotencoder = OneHotEncoder(categorical_features = [3])

#fit it to matrix X
X = onehotencoder.fit_transform(X).toarray()
```

```python
#avoiding the dummy variable trap (not necessarily required)

#just removed the first column from X (talking all rows and take all the column starting form index 1)
X = X[:,1:]
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.cross_validation import train_test_split

#1/3 of the dataset (10) will go in the test set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 1/3, random_state = 0)
```

```python
#fitting multiple linear regression to the training set

from sklearn.linear_model import LinearRegression

#creating object of LinearRegression class
regressor = LinearRegression()

#fit above object in our model
regressor.fit(X_train,Y_train)
```

```python
#predicting the test set results (creating a vector of predictions)
Y_pred = regressor.predict(X_test)
```

```python
#predicted profits
print(Y_pred)
#real profits
print(Y_test)
```

```python
#building the optimal model using Backward Elimination

#process involved in backward eliminattion
#select a significance level to stay in the model (usually 0.05)
#fit the full model with all possible predictions
#consider the predictor with highest P-values (probability). if P>SL, go one step down, otherwise finish the model
#remove the predictor
#fit model agagin without this variable

#it consist of including all the Independect variable (IV) at first then removing one by one to find optimal IV
#the optimal independent variable(IV) team IV will IV that have high effect on dependent variable
```

```python
import statsmodels.formula.api as sm

#add a column of 1 in the beginning of our matrix of feature X to include the constant be 0 in our MLR equation
#this is done to add a constant  b0 (column of all 1's) in the equation y = b0 + b1x1 + b2x2 + ..+bnXn
#in (50,1) 50 is the number of lines and 1 is the number of column
#axis = 0 is for adding it in a line and axis = 1 for a column

#here, we add a matrix of column 50 1's to X so its adds the column at the end but we need it in the beginning
# X = np.append(arr = X, values = np.ones((50,1)).astype(int), axis = 1)

#To add the column in beginning, we add the matrix X to the column of 50 1's
X = np.append(arr = np.ones((50,1)).astype(int), values = X, axis = 1)
```

```python
#creating optimal matirx of features

#Taking matirx of X with all its independent variable
X_opt = X[:,[0,1,2,3,4,5]]

#create a new regressor (object of new class of statsmodels library) and fit the regression to X_opt and Y
#endog = dependent vairable
#exog = array containing number of observations
regressor_OLS = sm.OLS(endog = Y, exog = X_opt).fit()
```

```python
#look for predictor with higher P (probability) value (table of p value; lower the p value, more significant IV will be)
regressor_OLS.summary()
```

```python
#in the table, const are all the 1's column we added(index 0), x1 and x2 are the dummy variables (index 1 and 2),
#and rests are other amounts spent

#remove index of 2 (since it has a p-value of 0.990 which is  > SL of 0.05)
X_opt = X[:,[0,1,3,4,5]]

regressor_OLS = sm.OLS(endog = Y, exog = X_opt).fit()
regressor_OLS.summary()
```

```python
#when removing over and over again, alwyas refer to original X feature matrix for getting index number

#remove index of 1 (since it has a p-value of 0.940 which is  > SL of 0.05)
X_opt = X[:,[0,3,4,5]]

regressor_OLS = sm.OLS(endog = Y, exog = X_opt).fit()
regressor_OLS.summary()
```

```python
#since we alreday removed the first two columns, which were the dummy varibles, the new x1, x2 and 3 in the above table are
#R&D spent, admin spent and marketing spent

#remove index of 4 (since it has a p-value of 0.602 which is > SL of 0.05)
X_opt = X[:,[0,3,5]]

regressor_OLS = sm.OLS(endog = Y, exog = X_opt).fit()
regressor_OLS.summary()
```

```python
#remove index of 5 (since it has a p-value of 0.060 which is > SL of 0.05)
X_opt = X[:,[0,3]]

regressor_OLS = sm.OLS(endog = Y, exog = X_opt).fit()
regressor_OLS.summary()
```
